```python
final = pd.DataFrame()

for j in range(1,51):
    url='http://books.toscrape.com/catalogue/page-{}.html'.format(j)
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url,headers=headers).text
    soup=BeautifulSoup(webpage,'lxml')
    elements = soup.find_all('article', class_='product_pod')
    for i in elements:
        name.append(i.find('h3').text.strip())
        price.append(i.find('p',class_='price_color').text.strip())
        ava.append(i.find('p',class_='instock availability').text.strip())
        rating.append(i.find('p', class_='star-rating').get('class')[1])


    d={'name':name,'price':price,'availability':ava,'rating':rating}
    df=pd.DataFrame(d)

    final=final.append(df,ignore_index=True)
```
